# 作业 5：图像数据的数组表示与操作

```python
import numpy as np
grayscale_image=np.random.randint(0,256,(200,300))
sepia_matrix = np.array([
    [0.393, 0.769, 0.189],
    [0.349, 0.686, 0.168],
    [0.272, 0.534, 0.131]
])
```

```python
image=np.stack([grayscale_image,grayscale_image,grayscale_image],axis=1)
image=np.clip(image, 0, 255, out=None)
print(image)
```

```python
sepia_image = np.clip(np.matmul(sepia_matrix,image),0,255)
print(sepia_image)
```

```python
a=np.split(sepia_image,3,axis=1)
print(a)
```
