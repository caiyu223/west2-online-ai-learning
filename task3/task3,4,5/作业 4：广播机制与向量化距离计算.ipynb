# 作业 4：广播机制与向量化距离计算

```python
import numpy as np
```

```python
points_A = np.random.randint(0,101,(5,2))
points_B = np.random.randint(0,101,(8,2))
```

```python
xa=points_A[:,0].reshape(5,1)
ya=points_A[:,1].reshape(5,1)
xb=points_B[:,0].reshape(1,8)
yb=points_B[:,1].reshape(1,8)
sx=(xa-xb)**2
sy=(ya-yb)**2
s=np.around(np.sqrt(sx+sy),2)
print(s)
sAmin=np.min(s,axis=1)
sBmin=np.min(s,axis=0)
index=np.where(sBmin<20)
print('points_B 中与 points_A 中至少一个点的距离小于 20 的所有点的索引:',index)
print(sAmin)
```
