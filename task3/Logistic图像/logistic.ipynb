```python
import numpy as np
import matplotlib.pyplot as plt

def logistic(x,L,k,x0,ax):
    y = L/(1+np.exp(-k*(x-x0)))
    ax.plot(x,y,label = f'L={L},k={k},x0={x0}')
    
```

```python
x = np.linspace(-10,10,1000)
fig,ax=plt.subplots(2,2,figsize=(15,15))
#fig = plt.figure(figsize=(80,80))
logistic(x,2,1,0,ax[0,0])
logistic(x,1,1,0,ax[0,0])
logistic(x,0.5,1,0,ax[0,0])
ax[0,0].legend()

logistic(x,1,2,0,ax[0,1])
logistic(x,1,1,0,ax[0,1])
logistic(x,1,0.5,0,ax[0,1])
ax[0,1].legend()

logistic(x,1,1,3,ax[1,0])
logistic(x,1,1,0,ax[1,0])
logistic(x,1,1,-3,ax[1,0])
ax[1,0].legend()


#plt.show()
```
